#Calibrating NIR instruments
___

##Problem description
*__Video: https://www.youtube.com/watch?v=K7Nd1H4iR68__*

FOSS is developing near infrared (NIR) spectroscopic equipment for measuring e.g. moisture and protein in food. In order to convert the NIR spectra into useful information mathematical models are applied to get reliable predictions of moisture, protein, etc. out. The development of these models is very resource demanding and hence it is important to be able to use the same mathematical models on more than one instrument. Due to measurement technology (Detector Diode Array (DDA), monochromator or interferometer) and the specific optical design of the instrument it is however difficult to use the same mathematical models on different instrument platforms.

__Typical differences in between instrument platforms are__

* Wavelength axis
* Spectral resolution and wavelength dependence of the resolution
* Sensitivity and nonlinearity of the response
* Wavelength dependent noise characteristics
A method for making spectral data from different NIR instrument platforms identical is therefore needed.

###The Challenge

__Instruments using three different NIR technologies will be available during the challenge:__

* Detector Diode Array (DDA)
* Monochromator
* Interferometer (FT-NIR)

__The challenge is to design a concept consisting of:__

1. A set of robust physical spectral standards, which can be measured on different NIR platforms. The number of standards being as low as possible (perhaps even zero).
2. A mathematical algorithm that will convert data from one platform to another, only by use of data from the physical spectral standards.

__The efficiency of the system will be tested by:__

1. Measuring the physical spectral standards defined by the concept on the three instrument platforms
2. Measuring a number of food samples on the same instruments
3. Transformation the food spectra to the same basis using the developed algorithms from the concept
4. Compare predictions from the transformed data using the current mathematical models for predicting moisture, protein, etc. from FOSS.
___

##Proposted solution

The basic idea is to use matrix factorization to construct a set of basic building blocks (components) that can work as a "common language" when translating between different platforms. These components is created by factorizing a matrix where each row vector is a concatenation of the measurements of a single food sample across all three instrument platforms.

The components constructed by matrix factorization techniques like PCA (Principal Component Analysis) or NMF (Non-negative Matrix Factorization) will also be highly interpretable and show a detailed relationship between the three platforms.

__Pros:__

* Interpretability (we can learn about the relationship between platforms)
* Easy translation between platforms through this "standardised common language".
* No need for forcing the different outputs into the same format (which will add noise).

__Cons:__

* Assuming linearity. (But this might acctually help regularizing which can be valuable when we do not have much data.)

###Loading in the data...

```python
# directory variables
data_dir = 'data/CroppedYale/'
file_path_pattern = data_dir + "yaleB{0:02d}/yaleB{0:02d}_P00A{1}.png"
```

```python
#imports
%matplotlib inline
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
def image_array(person, light_setting):
    file_path = file_path_pattern.format(person, light_setting)
    img = Image.open(file_path).convert("L")
    arr = np.asarray(img)
    return arr

def image_arrays(person, light_settings):
    return np.array([image_array(person, setting) for setting in light_settings])
        
example = image_arrays(person=5, light_settings=["+000E+45", "+060E+20", "-035E-20"])
img_shape = example[0].shape

print example
```

```python
def show_images(image_arrays, title=''):
    fig = plt.figure()
    fig.suptitle(title, fontsize=15)
    for i, array in enumerate(image_arrays):
        plt.subplot(1,len(image_arrays),i+1)
        fig = plt.imshow(array, cmap = matplotlib.cm.Greys_r)
        plt.title('Instrument %i' % (i+1))
        plt.axis('off')
        
show_images(example)
```

The input to the vector factorization is going to be the concatenation of all 3 measurements. Lets build a function to translate back and forth between image array and a single row.

```python
def images_to_row(array_of_arrays):
    return np.concatenate([np.concatenate(arrayarray) for arrayarray in array_of_arrays])

def row_to_images(array):
    return np.array([array.reshape(img_shape) for array in np.split(array, 3)]) #TODO: Hardcoded for now
    #reverse concatenation

print "Row shpae:", images_to_row(example).shape
print "Images shape:", row_to_images(images_to_row(example)).shape
```

Now the data can be loaded in as rows in a dataframe.

```python
def get_data(light_settings=["+000E+45", "+060E+20", "-035E-20"]):
    persons = [i for i in range(1, 39+1) if not i == 14] # no 14 does not exists
    return np.array([images_to_row(image_arrays(person, light_settings)) for person in persons])

faces = get_data()
```

```python
print "Size of dataframe: ", faces.shape
```

###Perform factorization
The idea is to construct a number of components that can be used to reconstruct new measurements. This set of components will work as a common language between the instruments.

*Code greatly inspired by: http://scikit-learn.org/stable/auto_examples/decomposition/plot_faces_decomposition.html#example-decomposition-plot-faces-decomposition-py*

```python
#imports
from time import time
from sklearn import decomposition
from sklearn.cluster import MiniBatchKMeans
import numpy.random as random
```

```python
#random.seed(0)
rng = random.RandomState(0)
n_components = 10
```

```python
random.shuffle(faces)
```

Some matrix factorization methods needs the data to be centered (0 mean).

```python
n_samples, n_features = faces.shape

# global centering
faces_centered = faces - faces.mean(axis=0)

# local centering
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)
```

```python
show_images(row_to_images(faces[0]), title='Non-centered')
show_images(row_to_images(faces_centered[0]), title='Centered')
```

Split data into training and test data

```python
test_size = 2

faces_training = faces[:-test_size]
faces_test = faces[-test_size:]

faces_centered_training = faces_centered[:-test_size]
faces_centered_test = faces_centered[-test_size:]
```

Do matrix factorization

```python
estimators = [
    {'name': 'Eigenfaces - RandomizedPCA',
     'estimator': decomposition.RandomizedPCA(n_components=n_components, whiten=True),
     'center': True},

    {'name': 'Non-negative components - NMF',
     'estimator': decomposition.NMF(n_components=n_components, init='nndsvda', beta=5.0,
                       tol=5e-3, sparseness='components'),
     'center': False},

    #{'name': 'Independent components - FastICA',
    # 'estimator': decomposition.FastICA(n_components=n_components, whiten=True),
    # 'center': True},

    #{'name': 'Sparse comp. - MiniBatchSparsePCA',
    # 'estimator': decomposition.MiniBatchSparsePCA(n_components=n_components, alpha=0.8,
    #                                 n_iter=100, batch_size=3),
    #                                 #random_state=rng),
    # 'center': True},

    {'name': 'MiniBatchDictionaryLearning',
     'estimator': decomposition.MiniBatchDictionaryLearning(n_components=15, alpha=0.1,
                                                  n_iter=50, batch_size=3,
                                                  random_state=rng),
     'center': True},

    {'name': 'Cluster centers - MiniBatchKMeans',
     'estimator': MiniBatchKMeans(n_clusters=n_components, tol=1e-3, batch_size=20,
                        max_iter=50, random_state=rng),
     'center': True},

    #{'name': 'Factor Analysis components - FA',
    # 'estimator': decomposition.FactorAnalysis(n_components=n_components, max_iter=2),
    # 'center': True},
]
```

```python
for estimator in estimators:
    print("Extracting the top %d %s..." % (n_components, estimator['name']))
    t0 = time()
    data = faces_training
    if estimator['center']:
        data = faces_centered_training
    estimator['estimator'].fit(data)
    train_time = (time() - t0)
    print("done in %0.3fs" % train_time)
    if hasattr(estimator['estimator'], 'cluster_centers_'):
        components_ = estimator['estimator'].cluster_centers_
    else:
        components_ = estimator['estimator'].components_
    #if hasattr(estimator['estimator'], 'noise_variance_'):
    #    plot_gallery("Pixelwise variance",
    #                 estimator['estimator'].noise_variance_.reshape(1, -1), n_col=1,
    #                 n_row=1)
    estimator['components'] = components_[:n_components]
    #plot_gallery('%s - Train time %.1fs' % (name, train_time),
    #components_[:n_components])
    #plt.show()
```

Lets see the 10 first PCA components:

```python
pca = estimators[0]
for i, component in enumerate(pca['components']):
    show_images(row_to_images(component), title="Component %i" % i)
```

This already teaches us a lot about the typical differences in our data. Like component 4, which shows how the nose of a face differs in different light settings.

###Reconstructing unseen data
Lets try to reconstruct an unseen sample and see how close we can get using the above 10 components.

First. Lets see what it looks like if we take 2/5 of the component 0 and 3/5 of component 1:

```python
def apply_weighting(components, weights):
    return np.sum((components.T * weights).T, axis=0)
```

```python
show_images(row_to_images(apply_weighting(pca['components'], [2,3,0,0,0,0,0,0,0,0])))
```

Alright. Lets try to reconstruct an unseen measurement.

To do this we need to define the reconstruction error that we optimize against:

```python
from scipy.optimize import minimize

def reconstruction_error(weights, unseen_data, components):
    return np.sum((unseen_data - apply_weighting(components, weights))**2)
```

```python
def make_reconstruction(unseen_data, components):
    x0 = np.array(np.zeros(len(components)))
    recon = minimize(reconstruction_error, x0, args=(unseen_data, components), method='BFGS', options={'disp': True})
    return recon

recon = make_reconstruction(unseen_data=faces_centered_test[0], components=pca['components'])
```

Lets compare the true data with the reconstruction:

```python
scikit_recon_x = pca['estimator'].transform(faces_centered_test[0])

show_images(row_to_images(faces_centered_test[0]), title='True unseen data')
show_images(row_to_images( apply_weighting(pca['components'], recon.x) ), title='Reconstruction (custom squared error)')
show_images(row_to_images( apply_weighting(pca['components'], scikit_recon_x) ), title='Reconstruction (scikit transform)')
```

```python
print scikit_recon_x
print recon.x / 10000000
```

###Reconstructing unseen data based on measurements from only one platform (or light setting when speaking of photos)
The challenge is to be able to translate between the different platforms (light settings), so lets use our 10 basic building blocks / components to do this. Lets pretend we get measurements from Instrument 3 and wants to predict what this looks like on instrument 1 an 2.

```python
measurement = row_to_images(faces_centered_test[0])[2]

def show_image(array):
    fig = plt.imshow(array.reshape(img_shape), cmap = matplotlib.cm.Greys_r)
    plt.title('Instrument %i' % (3))
    plt.axis('off')

print "Measurement:"
show_image(measurement)
```

We find the components weights that best reconstructs the measurement.

```python
components_3rd = pca['components'][:,-(n_features / 3):] # grap last 3rd of components
recon_pca = make_reconstruction(measurement.flatten(), components_3rd)
weights = recon_pca.x
```

And use these weights to reconstruct the measurment, but since we use the full length of the components we also get inferred what the measurement would look like on the two other platforms:

```python
recon_all = apply_weighting(pca['components'], weights)
```

```python
show_images(row_to_images(recon_all))
```

Thats it. Here we see that the result does not differ much from the case where all three measurements was used for creating the best reconstruction.

```python
print "Error using only measurement 3:", reconstruction_error(weights, faces_centered_test[0], pca['components'])
print "Error using all 3 measuremaents:", reconstruction_error(recon.x, faces_centered_test[0], pca['components'])
```
